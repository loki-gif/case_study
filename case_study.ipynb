```python
#Bank Account Manager
#Assume that initial balance is 1000 rs
balance = 1000
one_time_limit = 50000
owner = input("please enter the name of the Account Holder : ")
mob_no = int(input("please enter your registered mobile no : "))
acc_no = int(input("please enter the account no : "))

def check_balance() :
  global balance
  print(f"Name : {owner} Mobile no : {mob_no} Acc number : {acc_no}")
  print(f"Your balance is {balance}")

def deposit() :
  global balance,owner,mob_no,acc_no
  amount = int(input("please enter the amount you want to deposit : "))
  if amount <= one_time_limit :
    balance += amount
    print(f"Name : {owner} Mobile no : {mob_no} Acc number : {acc_no}")
    print(f"Your current balance is {balance}")
  else :
    print("The amount you are trying to deposit is too large !!!")

def withdraw() :
  global balance,owner,mob_no,acc_no
  amount = int(input("please enter the amount you want to withdraw : "))
  if amount <= balance :
    balance -= amount
    print(f"Name : {owner} Mobile no : {mob_no} Acc number : {acc_no}")
    print(f"Your current balance is {balance}")
  else :
    print("Insufficient balance")





```

```python
operation = input("Please enter the operation you want to do : ")
if operation == 'deposit':
  deposit()
elif operation == 'withdraw' :
  withdraw()
elif operation == 'check balance' :
  check_balance()
else :
  print("error")
```

```python
#Weather Data Analyzer
temp = []
hum = []

def add_temp () :
  a = int(input("enter the temperarute : "))
  temp.append(a)

def add_hum () :
  b = int(input("enter the humidity : "))
  hum.append(b)

def calc():
    # Avoid division by zero error if the lists are empty
    if len(temp) == 0 or len(hum) == 0:
        print("Not enough data to calculate.")
        return

    avg_temp = sum(temp) / len(temp)
    avg_hum = sum(hum) / len(hum)

    print(f"Average Temp: {avg_temp}, Average Humidity: {avg_hum}")

    # Weather Conditions logic
    if avg_temp > 30 and avg_hum > 60:
        print("It is Hot and Humid.")
    elif avg_temp > 30 and avg_hum <= 60:
        print("It is Hot and Dry.")
    elif avg_temp <= 30 and avg_hum > 60:
        print("It is Cool and Humid.")
    else:
        print("It is Cool and Pleasant.")
```

```python
inp = input("please enter which data you want to enter")
if inp == 'temperature':
  add_temp()
elif inp == 'humidity':
  add_hum()
```
