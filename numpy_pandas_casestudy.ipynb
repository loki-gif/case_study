4)  Check and remove duplicates

```python
data.duplicated()
```

```python
data = data.drop_duplicates(keep='first')
data
```

5)  Rename columns ("Title" > "Movie_Title")

```python
data.rename(columns={'Title' : 'Movie_Title'},inplace=True)
data
```

6)  Get unique genres

```python
data.Genre.unique()
```

7)  Extract number of genres per movie

```python
data["Genre_Count"] = data["Genre"].str.split(",").str.len()
data[["Movie_Title", "Genre", "Genre_Count"]]
```

8)  Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
data['Revenue (Millions)'] = data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].median())
data['Metascore'] = data['Metascore'].fillna(data['Metascore'].median())
data.isna().sum()
```

9)  Top 5 movies by Rating

```python
data.sort_values(by = ['Rating'],ascending = False,inplace = True)
data.head()
```

10) Movies with highest/lowest revenue

```python
data.nlargest(1,'Revenue (Millions)')
```

```python
data.nsmallest(100,'Revenue (Millions)')
```

11)  Avg. rating by Director

```python
dir_rate = data.groupby('Director')['Rating'].mean()
dir_rate
```

12) Movies per year

```python
movies_per_year = data.groupby('Year').size()

movies_per_year
```

13) Movies with rating > 8 & revenue > 100M

```python
movies = data[ (data['Rating'] > 8) & (data['Revenue (Millions)'] > 100)]
movies
```

14) Group by Director: avg Rating, Revenue

```python
dir = data.groupby('Director')[['Rating','Revenue (Millions)']].mean()
dir
```

15) Group by Year: count, avg Rating

```python
year = data.groupby('Year').agg(
    Movie_count = ('Movie_Title','count'),
    Avg_rating = ('Rating','mean')
  )
year
```

16) Create a new column Profitability:


a) Revenue divided by Duration


b) Find top 10 most profitable movies

```python
pd.Series([],name='Profitability')
data['Profitability']= data['Revenue (Millions)']/data['Runtime (Minutes)']
data.nlargest(10,'Profitability')
```

17) Create a column that categorizes movies by rating:


a) Rating < 6: Poor


b) 6 ≤ Rating < 7.5: Average

c) Rating ≥ 7.5: Excellent

```python
data['Feedback'] = data['Rating'].apply(
    lambda x : 'Poor' if x < 6
    else 'Average' if 6 <= x < 7.5
    else 'Excellent'

)
data
```

18) Create a custom function to score movies based on:


i)
Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score


Apply a function to dataset to assign the score

```python
def movie_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

data['Score'] = data.apply(movie_score, axis=1)

data
```
